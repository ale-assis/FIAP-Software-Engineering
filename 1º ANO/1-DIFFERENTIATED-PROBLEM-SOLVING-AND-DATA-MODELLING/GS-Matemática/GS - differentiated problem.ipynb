Resposta:  

Ao ajustar a curva, obtivemos a seguinte função:

f(x) = 4.99x^2 + -20289.99x + 20585084.93

```python
def plotar_funcao(func, x_range, titulo='', xlabel='x', ylabel='f(x)', legenda=''):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = func(x)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=legenda or titulo)
    plt.scatter(anos, animais, color='red', label='Dados observados')
    plt.title(titulo)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    if legenda:
        plt.legend()
    plt.show()

anos = np.array([2021, 2022, 2023, 2024])
animais = np.array([1190, 1155, 1030, 1015])

coeficientes = np.polyfit(anos, animais, 2)
modelo = np.poly1d(coeficientes)
a, b, c = coeficientes

print(f"Coeficientes da função ajustada: a={a}, b={b}, c={c}")

# Função ajustada
def funcao_ajustada(x):
    return a*x**2 + b*x + c

plotar_funcao(funcao_ajustada, [2021, 2024], 'População de animais pós diminuição de pesca predatória ao longo dos anos', 'Ano', 'Número de Animais', f'f(x) = {a:.2f}x^2 + {b:.2f}x + {c:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados fornecidos
T = np.array([2021, 2022, 2023, 2024])
N_T = np.array([1190, 1155, 1030, 1015])

# Ajustando um modelo polinomial de segundo grau (quadrático)
coeficientes = np.polyfit(T, N_T, 2)  # '2' significa um polinômio de grau 2
a, b, c = coeficientes  # Armazenando os coeficientes
print(f"Coeficientes ajustados: a = {a}, b = {b}, c = {c}")

# Criando o modelo polinomial com os coeficientes
modelo = np.poly1d(coeficientes)

# Gerando valores de T para a curva ajustada
T_ajustado = np.linspace(min(T), max(T), 100)
N_T_ajustados = modelo(T_ajustado)

# Plotando os dados e a curva ajustada
plt.figure(figsize=(10, 6))
plt.plot(T, N_T, 'o', label='Dados Originais')
plt.plot(T_ajustado, N_T_ajustados, 'r--', label=f'Curva Ajustada: {a:.2f}T^2 + {b:.2f}T + {c:.2f}')
plt.title('Número de Indivíduos ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Indivíduos')
plt.legend()
plt.grid(True)
plt.show()

plotar_funcao(funcao_ajustada, [2021, 2030], 'População de animais em 2030', 'Ano', 'Número de Animais', f'f(x) = {a:.2f}x^2 + {b:.2f}x + {c:.2f}')
```
