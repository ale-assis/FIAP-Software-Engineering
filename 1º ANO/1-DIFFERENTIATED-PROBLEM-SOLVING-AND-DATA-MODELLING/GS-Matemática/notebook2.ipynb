disco rígido atual do servidor, que tem 50tb? Responda em meses,
arredondando para baixo se for necessário.**

```python
max_storage_size_in_tb = 50
current_storage = 10
current_month = 0
month_percentage = 0.03

while current_storage <= max_storage_size_in_tb:
    current_month += 1
    current_storage += current_storage * month_percentage

print(f"Entre o mês {current_month - 1} e {current_month} ele ultrapassará o valor maximo de {max_storage_size_in_tb}TB com {current_storage:.0f}TB")
```

```python
preco_inicial = 50 #50 reais
vendas_iniciais = 800 # 800 ingressos
aumento_preco = 5 # aumento de 5 reais
reducao_vendas= 20 # redução de 20 vendas de ingresso
receita_desejada = 50000 
```

**3c (10 pontos) Para obter uma receita de 50000, qual seria o preço correto a
escolher?**

Resposta:   
Para obter uma receita de 50.000 reais, o preço ideal a ser cobrado é de 70 reais

```python
# calcular as vendas estimadas com base no preco 
def vendas_estimadas(preco):
    return vendas_iniciais - (preco - preco_inicial) // aumento_preco * reducao_vendas

# criar um array de precos possiveis
precos = np.arange(preco_inicial, 200, aumento_preco)

# calculando a receita para cada preco
receitas = vendas_estimadas(precos) * precos

# encontrando o preço que gera a receita desejada
preco_ideal = precos[np.argmin(np.abs(receitas - receita_desejada))]

print(f'Para obter uma receita de R$ {receita_desejada:.2f}, o preço ideal a ser cobrado é de R$ {preco_ideal:.2f}')
```

**3d (10 pontos) Para obter a receita máxima, qual seria o preço a escolher?**

Resposta:  
O preço do ingresso que maximiza a receita é de 125 reais, com a venda de 500 ingressos, resultando em uma receita de 62.500 reais

```python
preco_inicial = 50 
vendas_iniciais = 800 # 800 ingressos
aumento_preco = 5 # aumento de 5 reais
reducao_vendas= 20 # redução de 20 vendas de ingresso

# calcular preço e quantidade vendida
def calcular_preco_quantidade(n):
    preco = preco_inicial  + n * aumento_preco
    quantidade_vendas = vendas_iniciais - n * reducao_vendas
    return preco, quantidade_vendas

# calcular receita
def calcular_receita(n):
    preco, quantidade_vendas = calcular_preco_quantidade(n)
    return preco * quantidade_vendas

# calcular receita para diferentes valores de n
n_valores = np.arange(0, 41)  # Testar valores de n de 0 a 40
receitas = [calcular_receita(n) for n in n_valores] # list comprehensible com for que itera sobre cada elemento de n_valores

# encontrar o valor de n que maximiza a receita
n_maximo = n_valores[np.argmax(receitas)]
preco_maximo, quantidade_maxima = calcular_preco_quantidade(n_maximo)
receita_maxima = calcular_receita(n_maximo)

print(f'O preço do ingresso que maximiza a receita é R${preco_maximo:.2f}, com a venda de {quantidade_maxima} ingressos, resultando em uma receita de R${receita_maxima:.2f}.')
```
